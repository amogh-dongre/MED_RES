```python
import os
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray, rgba2rgb
from skimage.transform import resize
from skimage.feature import hog

base_dir = 'spectrograms'

image_size = (64, 64)

hog_features = []
labels = []

for class_folder in os.listdir(base_dir):
    class_path = os.path.join(base_dir, class_folder)
    if not os.path.isdir(class_path):
        continue
    label = int(class_folder.split('_')[-1])
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(class_path, img_file)
            img = imread(img_path)
            if img.shape[-1] == 4:
                img = rgba2rgb(img)
            gray_img = rgb2gray(img)
            resized_img = resize(gray_img, image_size)

            features = hog(resized_img,
                           orientations=9,
                           pixels_per_cell=(4, 4),
                           cells_per_block=(2, 2),
                           block_norm='L2-Hys',
                           visualize=False)

            hog_features.append(features)
            labels.append(label)

X = np.array(hog_features)
y = np.array(labels)

print("Shape of feature matrix X:", X.shape)
print("Shape of label array y:", y.shape)
```

```python
import pandas as pd
data = np.hstack((X, y.reshape(-1, 1)))
num_features = X.shape[1]
columns = [f'f{i}' for i in range(num_features)] + ['label']
df = pd.DataFrame(data, columns=columns)
df.to_csv('hog_features.csv', index=False)
print("CSV file 'hog_features.csv' has been created successfully.")
```

```python
df
```

```python
X = df.drop('label',axis=1)
y = df['label']
```

```python
X
```

```python
##We have a very large no.of features so lets perform dimensionality reduction using PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), 
         pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Explained Variance")
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_scaled)
```

```python
X_reduced
```

```python
num_components = X_reduced.shape[1]
pca_columns = [f'PC{i+1}' for i in range(num_components)]
X_reduced_df = pd.DataFrame(X_reduced, columns=pca_columns)
```

```python
X_reduced_df
```

```python
##Now let's select some features by ranking them using RandomForest Ranking
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_reduced_df, y)
importances = model.feature_importances_
feat_importances = pd.Series(importances, index=X_reduced_df.columns)
```

```python
feat_importances.sort_values(ascending=False)
```

```python
top_10_features = feat_importances.nlargest(10).index.tolist()
X_top10 = X_reduced_df[top_10_features]
```

```python
X_top10
```

```python
X_top10['target'] = y.reset_index(drop=True).astype(int)
X_top10
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='target', data=X_top10)
plt.title('Class Distribution')
plt.xlabel('Class Label')
plt.ylabel('Count')
plt.show()
```

```python
#As we can see the classes are imbalanced we have to use either Oversampling or Undersampling for this
from collections import Counter
print("Original Class Distribution:", Counter(X_top10['target']))
```

```python
X_features = X_top10.drop('target', axis=1)
y_labels = X_top10['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_labels, test_size=0.2, random_state=42, stratify=y_labels
)
```

```python
print("Before oversampling:", Counter(y_train))
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_train_res, y_train_res = rus.fit_resample(X_train, y_train)
print("After undersampling:", Counter(y_train_res))
```

```python
from imblearn.over_sampling import KMeansSMOTE, ADASYN
smote = KMeansSMOTE(sampling_strategy={1:300,0:300},k_neighbors=5,random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
print("After SMOTE:", Counter(y_train_smote))
```

```python
adasyn = ADASYN(n_neighbors=5,sampling_strategy={1:300,0:300},random_state=42)
X_train_adasyn, y_train_adasyn = adasyn.fit_resample(X_train, y_train)
print("After ADASYN:", Counter(y_train_adasyn))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier

estimators1 = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=10, random_state=42)),
    ('svc', SVC(probability=True, random_state=42)),
    ('xgb',XGBClassifier(n_estimators=100,learning_rate=0.3))
]

final_estimator1 = GradientBoostingClassifier()
stacking_clf1 = StackingClassifier(
    estimators=estimators1,
    final_estimator=final_estimator1,
    cv=10
)
stacking_clf1.fit(X_train_smote, y_train_smote)
y_pred1 = stacking_clf1.predict(X_test)
print("Stacking Accuracy(With SMOTE):", accuracy_score(y_test, y_pred1))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```
