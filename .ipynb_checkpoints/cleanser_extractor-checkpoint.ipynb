```python
import os
import numpy as np
import mne
from tqdm import tqdm
import scipy.stats as stats
import scipy.signal as signal
import pywt
import pandas as pd
```

```python
LOW_CUT = 0.5  
HIGH_CUT = 30.0  
SAMPLING_RATE = 256  
```

```python
def preprocess_eeg(file_path, save_path):
    #Loads an EDF file, applies a band-pass filter, and saves the preprocessed data.
    raw = mne.io.read_raw_edf(file_path, preload=True)
    raw.filter(LOW_CUT, HIGH_CUT, fir_design='firwin')
    
    data, times = raw.get_data(return_times=True)
    
    np.save(save_path, data)
    print(f"Processed and saved: {save_path}")
```

```python
def visualize_features(features_path):
    """Visualizes extracted features as a DataFrame and plots distributions."""
    features = np.load(features_path, allow_pickle=True).item()
    df = pd.DataFrame(features)
    
    print("Feature DataFrame:")
    print(df.head())
    
    df.plot(kind='box', figsize=(12, 6))
    plt.title("Feature Distributions")
    plt.show()
```

```python
input_dir = "./data"  
output_dir = "./preprocessed_data"
os.makedirs(output_dir, exist_ok=True)

# Processing  all EDF files finally!!!!!!!!!!!!!!!
edf_files = [f for f in os.listdir(input_dir) if f.endswith(".edf")]
for file in tqdm(edf_files, desc="Processing EEG Files"):
    input_path = os.path.join(input_dir, file)
    output_path = os.path.join(output_dir, file.replace(".edf", ".npy"))
    preprocess_eeg(input_path, output_path)
    
    data = np.load(output_path)
    features = extract_features(data)
    np.save(output_path.replace(".npy", "_features.npy"), features)

print("All files processed successfully!")
```

```python
def visualize_eeg(data_path, channel=0, sample_rate=SAMPLING_RATE):
    """Visualizes EEG data from a saved .npy file."""
    data = np.load(data_path)
    time_axis = np.arange(data.shape[1]) / sample_rate  # Convert sample indices to time in seconds
    
    plt.figure(figsize=(10, 4))
    plt.plot(time_axis, data[channel], label=f'Channel {channel}')
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title("EEG Signal Visualization")
    plt.legend()
    plt.show()
```

```python
if edf_files:
    sample_output = os.path.join(output_dir, edf_files[0].replace(".edf", ".npy"))
    visualize_eeg(sample_output)
    sample_features_output = sample_output.replace(".npy", "_features.npy")
    visualize_features(sample_features_output)
```

```python

def extract_features(data, sample_rate=SAMPLING_RATE):
    """Extracts time-domain, frequency-domain, and wavelet features from EEG data."""
    features = {}
    
    # Time-Domain Features
    features['mean'] = np.mean(data, axis=1)
    features['variance'] = np.var(data, axis=1)
    features['skewness'] = stats.skew(data, axis=1)
    features['kurtosis'] = stats.kurtosis(data, axis=1)
    features['rms'] = np.sqrt(np.mean(data**2, axis=1))
    features['zcr'] = np.sum(np.diff(np.sign(data), axis=1) != 0, axis=1) / data.shape[1]
    
    # Frequency-Domain Features (PSD & FFT)
    freqs, psd = signal.welch(data, fs=sample_rate, axis=1)
    features['spectral_centroid'] = np.sum(freqs * psd, axis=1) / np.sum(psd, axis=1)
    features['spectral_entropy'] = -np.sum(psd * np.log(psd + 1e-10), axis=1)
    features['peak_frequency'] = freqs[np.argmax(psd, axis=1)]
    
    # Time-Frequency Features (Wavelet Transform)
    wavelet = 'db4'
    coeffs = pywt.wavedec(data, wavelet, axis=1, level=3)
    features['wavelet_energy'] = [np.sum(np.square(c), axis=1) for c in coeffs]
    
    return features
```
